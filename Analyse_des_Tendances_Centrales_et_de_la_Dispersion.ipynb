```python
# 2. Chargement des données
df = pd.read_csv('Salary_Data.csv')
```

```python
# 3. Aperçu des données
print("Aperçu des premières lignes :")
print(df.head())

print("\nRésumé statistique :")
print(df.describe())

print("\nInformations sur les données :")
print(df.info())
```

```python
# 4. Calcul des mesures de tendance centrale
mean_age = df['YearsExperience'].mean()
median_age = df['YearsExperience'].median()
mode_age = df['YearsExperience'].mode()[0]
```

```python
mean_salary = df['Salary'].mean()
median_salary = df['Salary'].median()
mode_salary = df['Salary'].mode()[0]
```

```python
print(f"\nTendance centrale (Expérience) : Moyenne = {mean_age:.2f}, Médiane = {median_age}, Mode = {mode_age}")
print(f"Tendance centrale (Salaire) : Moyenne = {mean_salary:.2f}, Médiane = {median_salary}, Mode = {mode_salary}")
```

```python
# 5. Calcul des mesures de dispersion
std_age = df['YearsExperience'].std()
var_age = df['YearsExperience'].var()
range_age = df['YearsExperience'].max() - df['YearsExperience'].min()
```

```python
std_salary = df['Salary'].std()
var_salary = df['Salary'].var()
range_salary = df['Salary'].max() - df['Salary'].min()
```

```python
print(f"\nDispersion (Expérience) : Écart-type = {std_age:.2f}, Variance = {var_age:.2f}, Étendue = {range_age}")
print(f"Dispersion (Salaire) : Écart-type = {std_salary:.2f}, Variance = {var_salary:.2f}, Étendue = {range_salary}")
```

```python
# 6. Visualisations

# Histogrammes
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['YearsExperience'], kde=True, bins=10, color='skyblue')
plt.title("Distribution de l'expérience (années)")

```

```python
plt.subplot(1, 2, 2)
sns.histplot(df['Salary'], kde=True, bins=10, color='salmon')
plt.title("Distribution des salaires")
```

```python
plt.tight_layout()
plt.show()
```

```python
# Boîtes à moustaches
plt.figure(figsize=(12, 5))
```

```python
plt.subplot(1, 2, 1)
sns.boxplot(y=df['YearsExperience'], color='lightblue')
plt.title("Boxplot de l'expérience")
```

```python
plt.subplot(1, 2, 2)
sns.boxplot(y=df['Salary'], color='lightcoral')
plt.title("Boxplot du salaire")
```

```python
plt.tight_layout()
plt.show()
```

```python
# Nuage de points
plt.figure(figsize=(8, 6))
sns.scatterplot(x='YearsExperience', y='Salary', data=df)
plt.title("Expérience vs Salaire")
plt.xlabel("Années d'expérience")
plt.ylabel("Salaire")
plt.grid(True)
plt.show()
```

```python
# Scatter plot avec ligne de régression
sns.regplot(data=df, x='YearsExperience', y='Salary', line_kws={"color":"red"})
plt.title("Régression linéaire : expérience vs salaire")
plt.xlabel("Années d'expérience")
plt.ylabel("Salaire")
plt.show()
```
